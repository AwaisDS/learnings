```python
import torch
import torch.nn as nn
import torch.optim as optim

# -----------------------------
# 1. Create random input data
# -----------------------------
# We have 100 samples and 2 features in each sample
X = torch.randn(100, 2)

# Create labels manually:
# If sum of two features > 0, label = 1
# Otherwise label = 0
y = (X[:, 0] + X[:, 1] > 0).float().view(-1, 1)
```

```python
X
```

```python
# -----------------------------
# 2. Define a simple perceptron
# -----------------------------
class Perceptron(nn.Module):
    def __init__(self):
        super().__init__()

        # One neuron:
        # Input size = 2 features
        # Output size = 1 value
        self.linear = nn.Linear(2, 1)

    def forward(self, x):
        # Pass input through linear layer
        output = self.linear(x)

        # Apply sigmoid so output becomes between 0 and 1
        output = torch.sigmoid(output)

        return output
        
# Create model
model = Perceptron()
```

```python


# -----------------------------
# 3. Loss function and optimizer
# -----------------------------
loss_function = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

# -----------------------------
# 4. Train the model
# -----------------------------
for epoch in range(100):

    # Forward pass
    predictions = model(X)

    # Calculate loss
    loss = loss_function(predictions, y)

    # Clear old gradients
    optimizer.zero_grad()

    # Backpropagation
    loss.backward()

    # Update weights and bias
    optimizer.step()

    # Print loss every 10 epochs
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

```python
# -----------------------------
# 5. Test on a new sample
# -----------------------------
test_sample = torch.tensor([[2.0, 1.0]])

prediction = model(test_sample)

print("\nPrediction:", prediction.item())
```
