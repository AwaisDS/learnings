```python
 import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
df=fetch_california_housing()
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset
```

```python
 ## Independent features and dependent features

X=dataset
y=df.target
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
 ## standarizing the dataset

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test= scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
##cross validation
from sklearn.model_selection import cross_val_score
```

```python
r=LinearRegression()
r.fit(X_train,y_train)
```

```python
mse=cross_val_score(r,X_train,y_train ,scoring= 'neg_mean_squared_error',cv=10)
```

```python
np.mean(mse)
```

```python
#prediction 

r_pred = r.predict(X_test)
```

```python
import seaborn as sns
```

```python
sns.distplot(r_pred - y_test )
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(r_pred,y_test)
```

```python
score
```

```python
plt.scatter(y_test,r_pred)
```

```python
# Perfect prediction line
min_val = min(y_test.min(), r_pred.min())
max_val = max(y_test.max(), r_pred.max())

plt.scatter(y_test, r_pred)
plt.plot([min_val, max_val], [min_val, max_val]) 
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted (Multilinear Regression)")
plt.show()
```
